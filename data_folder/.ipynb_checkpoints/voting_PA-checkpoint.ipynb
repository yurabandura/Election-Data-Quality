```python
import csv
import geopandas as gp
import pandas as pd
import mysql.connector
```

```python
#Process presidential votes
state = 'PA'
democrat = 'trump'
republican = 'clinton'
STATE_FIPS = '42'

president = {}
reader = csv.reader(open('./election/2016-precinct-president.csv', 'r'))
for row in reader:
    if row[4] != state:
        continue
    #democr, republican
    id = row[13]
    
    id = id.zfill(5)
    county_fips = row[8]
    id = county_fips + id    
       
    
    if id in president.keys():
        votes = president[id]
    else:
        votes = [0, 0]
    if row[15] == democrat:
        votes[0] += int(row[21])
    if row[15] == republican:
        votes[1] += int(row[21])
    president[id] = votes
    
#print (president)
        
    
```

```python
file = "./PA/PA_precinct_c.geojson"
precincts = gp.read_file(file)
precincts.head()
```

```python
president_votes = precincts.copy()
president_votes = president_votes.drop(columns = ['type', 'centerX', 'centerY', 'canonical', 'geometry', 'neighbors', 'area'])

president_votes.insert(2, 'democrat', 0)
president_votes.insert(3, 'republican', 0)

president_votes.head()
```

```python
notFound = []
for i, row in president_votes.iterrows():
    id = str(row['precinct_c'])
    if id in president.keys():
        president_votes.at[i, 'democrat'] = president[id][0]
        president_votes.at[i, 'republican'] = president[id][1]
        #row['democrat'] = president[id][0]
        #row['republican'] = president[id][1]
    
        
    else:
        notFound.append(id)
    #print(president_votes.at[i, 'democrat'])
```

```python
print(notFound)
print(president_votes[(president_votes['democrat'] == 0) & (president_votes['republican'] == 0)].shape)
#for i, row in president_votes.iterrows():
    #if row['democrat'] == 0 or row['republican'] == 0:
       # print(i)
```

```python
#PROCESS CONGRESSIONAL VOTES 2016
#SENATE
STATE = 'PA'
DEMOCRATIC = 'mcginty'
REPUBLICAN = 'toomey'
senate2016 = {}

reader = csv.reader(open('./election/2016-precinct-senate.csv', 'r'))
for row in reader:
    if row[4] != STATE:
        continue
    #democr, republican
    id = row[13]
    id = id.zfill(5)
    county_fips = row[8]
    id = county_fips + id    
    
    if id in senate2016.keys():
        votes = senate2016[id]
    else:
        votes = [0, 0]
        
    if row[15] == DEMOCRATIC:
        votes[0] += int(row[21])
    if row[15] == REPUBLICAN:
        votes[1] += int(row[21])
    senate2016[id] = votes
    
#print (senate2016)
```

```python
#HOUSE
STATE = 'PA'
DEMOCRAT = ('balchunis' , 'boyle', 'brady', 'burkholder', 'cartwright', 'daugherty', 'doyle', 'evans',
            'halvorson', 'hartman', 'marsicano', 'mcclelland', 'molesevich', 'parrish', 'santarsiero', 'taylor',
           )
REPUBLICAN = ('barletta', 'connolly', 'costello', 'dent', 'fitzpatrick', 'jones', 'kelly', 'marino', 'mcallister',
              'meehan', 'murphy', 'perry', 'rothfus', 'shuster', 'smucker', 'thompson', 'williams', 
             )

house2016 = {}

reader = csv.reader(open('./election/2016-precinct-house.csv', 'r', encoding = "utf-8"))
for row in reader:
    if row[4] != STATE:
        continue
    #democrat, republican
    id = row[13]
    
    id = id.zfill(5)
    county_fips = row[8].zfill(3)
    id = county_fips + id    
    
    if id in house2016.keys():
        votes = house2016[id]
    else:
        votes = [0, 0]
        
    if row[15] in DEMOCRAT:
        votes[0] += int(row[21])
    elif row[15] in REPUBLICAN:
        votes[1] += int(row[21])
    house2016[id] = votes
    
#print (house2016)
```

```python
congress_votes2016 = precincts.copy()
congress_votes2016 = congress_votes2016.drop(columns = ['type', 'centerX', 'centerY', 'canonical', 'geometry', 'neighbors', 'area'])

congress_votes2016.insert(1, 'democrat', 0)
congress_votes2016.insert(2, 'republican', 0)

congress_votes2016.head()
```

```python
notFound = []
for i, row in congress_votes2016.iterrows():
    id = str(row['precinct_c'])
    
    if id in senate2016.keys():       
        congress_votes2016.at[i, 'democrat'] = senate2016[id][0]
        congress_votes2016.at[i, 'republican'] = senate2016[id][1]
    elif id in house2016.keys():
        democrats = congress_votes2016.at[i, 'democrat']
        congress_votes.at[i, 'democrat'] = democrats + house2016[id][0]
        republicans = congress_votes2016.at[i, 'republican']
        congress_votes2016.at[i, 'republican'] = republicans + house2016[id][1]
    else:
        notFound.append(id)
```

```python
#print(notFound)
print(congress_votes2016[(congress_votes2016['democrat'] == 0) & (congress_votes2016['republican'] == 0)].shape)
```

```python
pd.DataFrame(congress_votes2016).to_json("./election/PA_congress2016.json")
```

```python
### CONGRESS 2018 
STATE_FIPS = '42'
STATE = 'PA'
congress2018 = {}
reader = csv.reader(open('./election/2018_precinct_congress_PA.csv', 'r', encoding = "utf-8"))
for row in reader:
    if row[15] != STATE:
        continue
    #democrat, republican
    id = row[0].zfill(5)
    if row[6] in d.keys():
        id = STATE_FIPS + d[row[6]].zfill(3) + id
        
    else:        
        continue
    if id in congress2018.keys():
        votes = congress2018[id]
    else:
        votes = [0, 0]
        
    if row[2] == 'democrat':
        votes[0] += int(row[4])
    elif row[2] == 'republican':
        votes[1] += int(row[4])
    congress2018[id] = votes
    
#print (congress2018)
```
